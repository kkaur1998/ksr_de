```python
#self_join
data = [
[1, 'RAMAN', 5],
[2, 'SARAB', 5],
[3, 'PARAM', 5],
[4, 'KAMAL', 5],
[5, 'SANTHOSH', 6],
[6, 'MAHESH', np.nan]]
columns = ['id','name','manager']
emp = pd.DataFrame(data,columns=columns)
```

```python
emp1=emp.copy()
```

```python
emp.merge(emp,left_on='manager',right_on='id',how='left')[['id_x','name_x','name_y']].rename({'id_x':'id','name_x':'name','name_y':'manager_name'})
```
