```python
#10)Write a query to calculate total sales, total profit and total order quantity?
df2[['Sales','Profit','OrderQuantity']].sum()
```

```python
#11)Write a query to calculate the total sales amount, total order quantity for each category. 
#Display the category, total sales, and total order qty and order by total sales from 
#highest to lowest?
df2[['Category','Sales','OrderQuantity']].groupby('Category',as_index=False).sum().sort_values('Sales',ascending=False)
```

```python
# 12)Write a query to calculate the total profit amount for each category. Display the 
# category, total profit, and total order qty and order by total profit from highest to lowest?

df2[['Category','Profit','OrderQuantity']].groupby('Category',as_index=False).sum().sort_values('Profit',ascending=False)
```

```python
# 13)Write a query to fetch the subcategories where total sales are greater than 100000?

df2[['SubCategory','Sales']].groupby('SubCategory',as_index=False).sum().query('Sales>100000')
```

```python
# 14)Write a query to fetch the products where total profit is greater than 2500 and sort it 
# based on profit from highest to lowest??

df2[['ProductName','Profit']].groupby('ProductName',as_index=False).sum().query('Profit>2500').sort_values('Profit',ascending=False)
```

```python
k = df2[['ProductName','Profit']].groupby('ProductName',as_index=False).sum()
k[k['Profit']>2500].sort_values('Profit',ascending=False)
```

```python
#15)Write a query to get the total sales and total profit for OƯice Supplies category?
df2[df2['Category'] == 'Office Supplies'][['Category','Sales','Profit']].groupby('Category',as_index=False).sum()
```

```python
# 16)Write a query to get the total sales and total profit for Furniture category and Tables, 
# Bookcases sub-categories ?
df2[(df2['Category'] == 'Furniture') & (df2['SubCategory'] == 'Bookcases')][['Category','Sales','Profit']].groupby('Category',as_index=False).sum()
```

```python
# 17)Write a query to get the total sales and total profit for Technology category and the 
# Accessories, Copiers, Phones sub-categories ?

df2[(df2['Category'] == 'Technology') & (df2['SubCategory'].isin(['Accessories','Copiers','Phones']))][['Category','Sales','Profit']].groupby('Category',as_index=False).sum()
```

```python
#taking it for further calculations from both tables
df3 = df1.merge(df2,how='left',on='OrderID')
```

```python
# 18)Write a query to get total sales and total profit by Region, Segment and sort the sales 
# from highest to lowest?? 
df3[['Region','Segment','Sales','Profit']].groupby(['Region','Segment']).sum().sort_values(['Region','Sales'],ascending=False)
```

```python
# 19)Write a query to get total sales and total profit by Country, State and city and sort the 
# sales from highest to lowest?? 
df3[['Country','State','City','Sales','Profit']].groupby(['Country','State','City'],as_index=False).sum().sort_values('Sales',ascending=False)
```

```python
# 20)Write a query to get total sales and total orderqty by CustomerName sort it by total 
# sales from highest to lowest? 

df3[['CustomerName','Sales','OrderQuantity']].groupby('CustomerName',as_index=False).sum().sort_values('Sales',ascending=False)
```

```python
# 21)Identify the top 5 products with the highest sales (by sales amount). Show the product 
# name, total sales, and total qty?

df3[['ProductName', 'Sales', 'OrderQuantity']].groupby('ProductName').sum().sort_values('Sales',ascending=False).head(5)
```

```python
#22)Write a query to get total sales by City having sales greater than 35000? 
df3[['City','Sales']].groupby('City').sum().sort_values('Sales',ascending=False).query('Sales>35000')
```

```python
# 23)Write a query to get total sales by CustomerName having sales greater than 10000? 
df3[['CustomerName','Sales']].groupby('CustomerName').sum().query('Sales>10000')
```

```python
# 24)Write a query to get total sales and total profit by shipmode and sort the sales in 
# ascending order and profit in descending order? 

df3[['ShipMode','Sales','Profit']].groupby('ShipMode').sum().sort_values(['Sales','Profit'],ascending=[True,False])
```

```python
# 25)Write a query to get total sales for North and central region?
df3[df3['Region'].isin(['North','Central'])][['Region','Sales']].groupby('Region').sum()
```

```python
# 26)Write a query to get total sales and total profit for Italy and Spain countries? 
df3[(df3['Country']=='Italy') | (df3['Country']=='Spain')][['Country','Sales','Profit']].groupby('Country',as_index=False).sum()
```

```python
# 27)Write a query to get the total sales and total profit for each year sort the sales from 
# highest to lowest?
from datetime import datetime
df3['year'] = df3['OrderDate'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d').year)
df3[['year','Sales','Profit']].groupby('year', as_index=False).sum().sort_values('Sales',ascending=False)
```

```python
# 28) Find the top 10 customers who spent the most across all transactions. Display the 
# customer name, total amount spent, and number of orders placed?

k = df3[['CustomerName','Sales','OrderID']].groupby('CustomerName',as_index=False).agg({'OrderID':'count','Sales':'sum'}).sort_values('Sales',ascending=False)
k.head(10).reset_index(drop=True).rename(columns={'OrderID':'Total_orders_placed','Sales':'Total_amount_spent'})
```

```python
# 29)Write a query to find which products are most preferred by customers based on the 
# total sales. Display customer name, favorite product9 (top 3 products per each customer), 
#and total sales on that product?

k2 = df3[['CustomerName', 'ProductName', 'Sales']].groupby(['CustomerName', 'Sales']).sum().sort_values(['CustomerName','Sales'],ascending=False)
k2.groupby('CustomerName').head(3)
```

```python
#30)Write a query to get 7th rank customer name based on total sales? Display customer 
# name, sales amount and rank.
df3[['CustomerName','Sales']].groupby('CustomerName',as_index=False).sum().sort_values('Sales',ascending=False).reset_index(drop=True).iloc[7,:]
```

```python
# 31)Write query to get total sales and total profit in years 2011 and 2013 Display year wise 
# total sales and total profit?
df3[df3['year'].isin([2011,2013])][['year','Sales','Profit']].groupby('year',as_index=False).sum()
```

```python
# 32)Write a query to get total sales, total profit and total order qty by country, state, category 
# and sub-category and sort it from highest to lowest based on sales column?

df3[['Country','State','Category','SubCategory','Sales','Profit','OrderQuantity']].\
groupby(['Country','State','Category','SubCategory']).sum().sort_values('Sales',ascending=False)
```

```python
# 33)write a function to get the region sales dynamically, if we pass any region, that region 
# sales should be displayed?
def sales_by_region(region):
    return df3[df3['Region']==region][['Region','Sales']].groupby('Region',as_index=False).sum()

reg = input("enter region here: ").capitalize()
sales_by_region(reg)
```

```python
# 34)write a function to get the country sales, profit dynamically, if we pass any country, that 
# country sales, profit should be displayed?

def sales_by_country(country):
    k3 = df3[df3['Country']==country]
    return k3[['Country','Sales','Profit']].groupby('Country',as_index=False).sum()

country = input("enter country here: ").capitalize()
sales_by_country(country)
```

```python
# 35)write a function to get the Category orderquantity dynamically, if we pass any Category, 
# that Category orderqunatity should be displayed.?

def orderqty_by_category(category):
    k4 = df3[df3['Category']==category]
    return k4[['Category','OrderQuantity']].groupby('Category', as_index=False).sum()
category = input("enter category here: select from : 'Office Supplies', 'Furniture', 'Technology'").capitalize()
orderqty_by_category(category)
```

```python
# 36)write a function to get the City sales and profit dynamically, if we pass any city, that City 
# sales and profit should be displayed

def sales_by_city(city):
    k4 = df3[df3['City']==city]
    return k4[['City','Sales']].groupby('City', as_index=False).sum()
city = input("enter city here: ").capitalize()
sales_by_city(city)
```
